```python
def kmeans_cluster(data, w):
    n_features = data.shape[0]
    n_data = data.shape[1]
    m = w.shape[1]
    m_arr = np.zeros(n_data)
    for j in range(n_data):
        dst_arr = np.zeros(m)
        for k in range(m):
            dst_arr[k] = np.linalg.norm(data[:, j] - w[:,k])
        m_arr[j] = np.argmin(dst_arr)
    return m_arr
```

Exercise H8.2: Online K-means Clustering

```python
def online_kmeans(data, num_m, tau=0.99, eps0=0.1):
    n_features = data.shape[0]
    n_data = data.shape[1]

    w = np.ones((n_features, num_m))
    w = np.multiply(w.T, np.mean(data, axis=1)).T + np.random.normal(scale=1e-3, size=w.shape)
    
    tau_multiplier = np.ones(n_data)
    tau_multiplier[int(n_data*0.25):] = tau

    W = np.zeros((n_data+1, n_features, num_m))
    W[0] = w

    eps = eps0*tau_multiplier[0] 

    for i in range(n_data):
        dst_arr = np.zeros(num_m)
        for k in range(num_m):
            dst_arr[k] = np.linalg.norm(data[:, i] - w[:,k])
        m_min = np.argmin(dst_arr)

        w[:,m_min] += eps*(data[:,i] - w[:,m_min])
        
        W[i+1,:,:] = w    
        
        eps *= tau_multiplier[i]

    return W
```

```python
def online_kmeans_costs(data, w):
    m = kmeans_cluster(data, w)
    return cost(w, m, data)
```

```python
num_m = 4
W = online_kmeans(data, num_m)
w = W[-1,:,:]
m = kmeans_cluster(data, w)
```

Exercise H8.3: Soft K-means Clustering

```python
def soft_kmeans(data, num_m, beta, theta = 1e-3):
    n_features = data.shape[0]
    n_data = data.shape[1]

    W = np.ones((n_features, num_m))
    W = np.multiply(W.T, np.mean(data, axis=1)).T + np.random.normal(scale=1e-2, size=W.shape)
    W_init = W.copy()
    
    del_w = np.ones_like(W)
    a_probs = np.zeros((num_m, n_data))

    while(del_w > theta).any():
        for k in range(num_m):
            for a in range(n_data):
                a_probs[k, a] = np.exp(-0.5 * beta * np.linalg.norm(data[:,a] - W[:,k])**2)
                
        a_probs_sum = np.sum(a_probs, axis=0)
        a_probs = a_probs / a_probs_sum

        new_W = np.zeros_like(W)
        for k in range(num_m):
            sum_w = np.sum(np.multiply(data, a_probs[k,:]), axis=1)
            new_W[:, k] = sum_w / np.sum(a_probs[k,:])

        del_w = np.abs(new_W - W)
        W = new_W
    return W, W_init
```

```python
betas = np.arange(0.2, 2.2, 0.2)
num_m = 12
for beta in betas:
    w, w_init = soft_kmeans(data, num_m, beta=beta)
    m_arr = kmeans_cluster(data, w)
    for i in range(num_m):
        idx = m_arr == i
        m_data = data[:,idx]
        plt.scatter(m_data[0], m_data[1], label=i, color=colors[i])
    plt.title('Soft K-Means for Beta={}'.format(beta))
    plt.scatter(w[0,:], w[1,:], color='olive')
    plt.scatter(w_init[0,:], w_init[1,:], color='palegoldenrod')
    plt.show()
```

```python
def soft_kmeans_annealing(data, num_m, beta0=0.2, tau=1.1, theta=1e-3,num_iter=10):
    n_features = data.shape[0]
    n_data = data.shape[1]

    W_arr = np.ones((num_iter, n_features, num_m))
    W = np.ones((n_features, num_m))
    W = np.multiply(W.T, np.mean(data, axis=1)).T + np.random.normal(scale=1e-2, size=W.shape)
    

    beta=beta0
    for i in range(num_iter):
        del_w = np.ones_like(W)
        a_probs = np.zeros((num_m, n_data))

        while(del_w > theta).any():
            for k in range(num_m):
                for a in range(n_data):
                    a_probs[k, a] = np.exp(-0.5 * beta * np.linalg.norm(data[:,a] - W[:,k])**2)
                    
            a_probs_sum = np.sum(a_probs, axis=0)
            a_probs = a_probs / a_probs_sum

            new_W = np.zeros_like(W)
            for k in range(num_m):
                sum_w = np.sum(np.multiply(data, a_probs[k,:]), axis=1)
                new_W[:, k] = sum_w / np.sum(a_probs[k,:])

            del_w = np.abs(new_W - W)
            W = new_W

        beta *= tau
        W_arr[i,:,:] = W

    return W_arr
```

```python
num_m = 12
W = soft_kmeans_annealing(data, num_m, beta0=0.2, num_iter=50)
m_arr = kmeans_cluster(data, W[-1])
```

```python
for i in range(num_m):
    idx = m_arr == i
    m_data = data[:,idx]
    plt.scatter(m_data[0], m_data[1], label=i, color=colors[i])
plt.title('Soft K-Means with Annealling'.format(beta))
plt.scatter(W[-1,0,:], W[-1,1,:], color='olive')
plt.scatter(W[0, 0,:], W[0, 1,:], color='palegoldenrod')
plt.show()
```
